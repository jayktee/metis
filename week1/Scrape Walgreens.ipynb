```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.walgreens.com/storelocator/find.jsp?requestType=locator&state=NY&city=NEW%20YORK&from=localSearch')
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
results = soup.find_all("li", class_="card card__store-listing")
results
```

```python
for link in results.find_all('a'):
    print(link.get('href'))
```
