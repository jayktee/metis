## Exercise 1

```python
one = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_160903.txt")
two = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_160910.txt")
three = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_160917.txt")
```

```python
df = pd.concat([one, two, three], join="inner", ignore_index=True)
```

```python
df.head()
```

## Exercise 2

First I will convert the column names to lowercase and remove any special characters, to make it easier to clean.

```python
df.columns = [x.lower().replace('/','_') for x in df.columns]
df.head()
```

```python
df.dtypes
```

Pandas assigned an `object` (i.e. string) type to the `DATE` and `TIME` columns on import. I will convert these columns to a single date-time column.

```python
df['date_time'] = pd.to_datetime(df.date + ' ' + df.time)
df.head()
```

```python
df.dtypes
```

## Exercise 3

```python
df.groupby(['c_a', 'unit', 'scp', 'station']).head()
```

It looks like for a unique combination of the first 4 columns (i.e. a unique turnstile), each row represents a 4-hour time window on a specific date with the total number of entries and exits within that window. Therefore, to calculate the total number of entries and exits within a day, we would have to sum all the entries and exits across the 4-hour windows within a specific date.

To get the max number of entries per day, we group by each unique turnstile and the date, and use the `.agg()` method to get the `max` per day.

```python
df.groupby(['c_a', 'unit', 'scp', 'station', 'date']).agg({'entries':'max'})
```
